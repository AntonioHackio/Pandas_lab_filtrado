```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)

df_union_final = pd.read_csv("datos/df_union_final.csv", index_col = 0)
```

```python
df_union_final
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
df_Movie = df_union_final.loc[df_union_final.type == "Movie"]
df_2021 = df_Movie[df_Movie.release_year == 2021] 
df_2021
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
df_90 = df_union_final.loc[df_union_final.duration == "90 min"]
df_90[["title", "duration"]]
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
df_india = df_union_final.loc[df_union_final.country == "India"]
df_india_2018 = df_india.loc[df_india.release_year >= 2018]
df_india_2018
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
condicion_TV14 = df_union_final["rating"] == "TV-14"
condicion_R = df_union_final["rating"] == "R"

df_rating = df_union_final[condicion_TV14 | condicion_R]
df_rating
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos ha dirigido cada uno de los directores?

```python
filtro_directores = ["Christopher Nolan", "Martin Scorsese"]

df_directores = df_union_final[df_union_final["director"].isin(filtro_directores)]

count_nolan = df_directores["director"].value_counts()
print(count_nolan)
df_directores
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
filtro_titulos = ["United States", "United Kingdom", "Canada"]

df_titulos = df_union_final[df_union_final["country"].isin(filtro_titulos)]
df_titulos.value_counts("Language")
#Dos películas.
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python
filtro_fecha = [2015, 2016, 2017, 2018, 2019, 2020]

df_fecha = df_union_final[df_union_final["release_year"].isin(filtro_fecha)]
df_fecha
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
df_puntuacion = df_union_final[df_union_final["IMDB Score"].between(6.5, 8.0, inclusive = "both" )]
df_puntuacion
```

#### Ejercicio 9: ¿Cuáles son las películas cuya duración de 90 y 150 minutos?

```python
df_duracion = df_union_final.query('duration == "90 min" or duration == "150 min"') 
df_duracion
```

#### Ejercicio 10: ¿Cuáles son los títulos que contienen la palabra 'Love' en su título y que sean de tipo TV Show?

```python
df_love_filter = df_union_final[(df_union_final['title'].str.contains('Love')) & (df_union_final['type'] == 'TV Show')]
df_love_filter
```

#### Ejercicio 11: ¿Cuáles son los títulos en los que el director tiene el apellido 'Johnson'?

```python
df_apellido = df_union_final[(df_union_final['director'].str.contains('Johnson', na = False))]
df_apellido
```

#### Ejercicio 12: ¿Tenemos algún documental ('Documentaries') de Mike Tyson?

#### Ejercicio 14: Filtra las columnas que contienen la palabra 'date'.

#### Ejercicio 15: Usando `query()`, filtra todos los títulos del país 'France' con una duración de 80 minutos.

```python
df_15 = df_union_final.query('duration == "80 min" and country == "France"') 
df_15
```

#### Ejercicio 16: Filtra los títulos que son películas ('Movie') y fueron lanzados antes de 2010.

```python
df_16 = df_union_final.query('type == "Movie" and  release_year < 2010')
df_16
```

#### Ejercicio 17: Encuentra los 10 títulos con las puntuaciones más altas en IMDb.

```python
df_top10 = df_union_final.nlargest(10, "IMDB Score")
df_top10
```

#### Ejercicio 18: Encuentra los 5 títulos con las puntuaciones más bajas en IMDb.

```python
df_down5 = df_union_final.nsmallest(5, "IMDB Score")
df_down5
```

#### Ejercicio 19: Filtra todos los títulos que no tienen un director asignado (valores nulos en la columna 'director').

```python
df_null_director = df_union_final[df_union_final["director"].isnull()]
df_null_director
```

#### Ejercicio 20: Filtra todos los títulos que tienen información completa en las columnas 'title', 'country' y 'IMDB Score'.

```python
df_completos = df_union_final[df_union_final["title"].notnull() & df_union_final["country"].notnull() & df_union_final["IMDB Score"].notnull()]
df_completos
```
